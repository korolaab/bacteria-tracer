```python
import cv2 # video and image
import csv # txt,csv,tsv files 
import numpy as np# matrices
import matplotlib.pyplot as plt
```

```python
bacteria_coordinates=[]
with open("Coli/100001.txt","r") as f:
    reader = csv.reader(f,delimiter=':')
    for row in reader:
        bacteria_coordinates.append([float(row[0]),float(row[1])])


#load coordinates
```

```python
frame=cv2.imread("/home/korolaab/works/bacteria_tracer/Coli_frames/im_001.png",cv2.IMREAD_UNCHANGED) #load frame
bac_coord = np.array(bacteria_coordinates)# load coord to matrix
bac_coord = bac_coord.astype("int32") # convert to integer
for i in range(bac_coord.shape[0]):
    cv2.circle(frame,center=(bac_coord[i,0],bac_coord[i,1]),radius=3,color=(255,0,0),thickness=-1)
```

```python
import os
```

```python
txt_files=os.listdir("Coli/")
frames_files=os.listdir("Coli_frames/")
```

```python
txt_files.sort()
frames_files.sort()
```

```python
del frames_files[0]
```

```python
for i in range(len(txt_files)):
    with open("Coli/"+txt_files[i],"r") as f:
        reader = csv.reader(f,delimiter=':')
        bacteria_coordinates=[]
        for row in reader:
            bacteria_coordinates.append([float(row[1]),float(row[0])])
    frame=cv2.imread("Coli_frames/"+frames_files[i],cv2.IMREAD_UNCHANGED) #load frame
    bac_coord = np.array(bacteria_coordinates)# load coord to matrix
    bac_coord = bac_coord.astype("int32") # convert to integer
    for k in range(bac_coord.shape[0]):
        cv2.circle(frame,center=(bac_coord[k,0],bac_coord[k,1]),radius=3,color=(255,0,0),thickness=-1)
    cv2.imwrite("dataset/frames_with_marks/%s.png"%(i),frame)
    
```
